```python
import dill
import pandas as pd
```

```python
med = pd.read_csv('/home/qlunlp/ylh/database/eICU/治疗/medication.csv', usecols=['drugname'])
unique_drugname = med['drugname'].astype(str).unique().tolist()
unique_drugname = [name.lower() for name in unique_drugname]
len(unique_drugname), unique_drugname
```

```python
rxnorm = pd.read_csv('/home/qlunlp/ylh/database/drug/rxnorm_mappings.txt', sep='|', usecols=['RXSTRING'])
unique_rxnorm = rxnorm['RXSTRING'].unique().tolist()
# 转小写
unique_rxnorm = [name.lower() for name in unique_rxnorm]

# 只取第一个单词，并且第一个单词是字母类型
unique_rxnorm = list(set([name.split(' ')[0] for name in unique_rxnorm if name and name.split(' ')[0].isalpha()]))
len(unique_rxnorm), unique_rxnorm
```

```python
# 寻找同时出现在两个列表中的药物名称
common_drugs = set(unique_drugname).intersection(set(unique_rxnorm))
len(common_drugs), common_drugs
```
